```python
input_file = "paste.txt"

with open(input_file, "r") as f:
    # Skip header (first line), if present
    lines = f.read().splitlines()

# Remove header if needed
if not lines[0].replace('.', '', 1).isdigit():
    lines = lines[1:]

comma_str = ",".join(lines)
# Example: Save a single comma-separated string to a text file
with open("output.txt", "w") as f:
    f.write(comma_str)
```

```python
with open("output.txt", "r") as f:
    content = f.read()

# Convert string to list of floats
values = [float(x) for x in content.strip().split(",") if x]

import numpy as np
std_dev = np.std(values, ddof=1)  # ddof=1 gives sample std. dev (same as pandas .std())
print("Standard deviation:", std_dev)
```
