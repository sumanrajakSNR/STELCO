```python
import pandas as pd
import os
```

```python
folder_path = r"D:\Stelco\Work\Dynamic Correlation\Key\Master whole Phase"  # UPDATE this to your folder path

excel_path = r"D:\Stelco\Work\Dynamic Correlation\QVW Data for S&R 2024-7-2025 - 1008GCQ.xlsx"  # UPDATE this to your file path
sheetname = "All Data"
```

```python
file_names = os.listdir(folder_path)
folder_pickle_ids = []
for i in file_names:
    folder_pickle_ids.append(i.split(".")[0])

folder_pickle_ids_set = set(folder_pickle_ids)

# Step 2: Read Excel file and clean 'Pickle ID' column
df_excel = pd.read_excel(excel_path, sheet_name=sheetname,converters={'Pickle ID': lambda x: str(int(x)).strip().lstrip('0') if pd.notna(x) else ''})

df_excel = df_excel[df_excel['Pickle ID'].notna()]
df_excel_set = set(df_excel['Pickle ID'])


# print(df_excel_set)
missing_coils = df_excel_set.difference(folder_pickle_ids_set)


# Step 3: Filter rows where Pickle ID is NOT present in folder file names
missing_df = df_excel[~(df_excel['Pickle ID'].isin(folder_pickle_ids_set))]

# Step 4: Save result to CSV
output_path = r"D:\Stelco\Work\Dynamic Correlation\Missing_Coils.csv"
missing_df.to_csv(output_path, index=False)

print(f"✅ Saved {len(missing_df)} missing Pickle ID rows to '{output_path}'")
```
