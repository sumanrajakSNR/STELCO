```python
import pandas as pd
import os

# Input and output folder paths
coil_folder = r"D:\Stelco\Work\Dynamic Correlation\Key\Whole_Phase_Good_Phased"
output_folder = r"D:\Stelco\Work\Dynamic Correlation\Key\Whole_Phase_Good_Clustered"
master_csv = r"Anomaly CSV\Good Coils\coil_anomaly_isolationforest_clustering_data_merged.csv"

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Columns to merge from master CSV
columns_to_merge = ['K means Cluster']

# Read master CSV without setting index
master_df = pd.read_csv(master_csv)

# Process each coil file
for filename in os.listdir(coil_folder):
    if filename.endswith('.csv'):
        coil_id = os.path.splitext(filename)[0]
        file_path = os.path.join(coil_folder, filename)
        
        # Filter rows matching coil_id in master_df
        matching_rows = master_df[master_df['Coil ID'] == int(coil_id)]
        if not matching_rows.empty:
            values = matching_rows.iloc[0][columns_to_merge]
            df = pd.read_csv(file_path)
            for col in columns_to_merge:
                df[col] = values[col]
            # Save updated CSV in output folder with same filename
            new_path = os.path.join(output_folder, filename)
            df.to_csv(new_path, index=False)
```
