```python
output_csv = 'good_and_bad_coils_10_meter_sampling_combined.csv'
write_header = True
```

```python
for csv_file in csv_files:
    df = pd.read_csv(csv_file,low_memory=False)
    # selected_columns = df.columns
    batch_rows = []

    for coil_id, group in df.groupby('Coil ID'):
        group_sorted = group.sort_values('Coil Length [30ms]').reset_index(drop=True)
        start = group_sorted['Coil Length [30ms]'].min()
        stop = group_sorted['Coil Length [30ms]'].max()
        desired_lengths = np.arange(start, stop + 10, 10)
        
        indices = get_nearest_indices(group_sorted['Coil Length [30ms]'].values, desired_lengths)
        unique_indices = np.unique(indices)
        
        # Select only desired columns
        batch_rows.append(group_sorted.iloc[unique_indices][selected_columns])

    # Concatenate all coils for this file
    if batch_rows:
        batch_df = pd.concat(batch_rows, ignore_index=True)
        batch_df.to_csv(output_csv, mode='a', header=write_header, index=False)
        write_header = False  # Only write header for the first batch

print("All files appended to:", output_csv)
```
