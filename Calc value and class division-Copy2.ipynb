```python
def load_class_division_csv(class_division_csv, class_cols):
    df = pd.read_excel(class_division_csv, sheet_name="All Data")
    df['Pickle ID'] = df['Pickle ID'].astype(str).str.split('.').str[0].str.strip()

    # Convert Coil_Prod_Date to datetime
    df['Coil_Prod_Date'] = pd.to_datetime(df['Coil_Prod_Date'], errors='coerce')
    
    # Filter rows for dates between April 1, 2024 and April 30, 2025 (assuming inclusive)
    start_date = pd.Timestamp('2024-04-01')
    end_date = pd.Timestamp('2025-04-30')
    df = df[(df['Coil_Prod_Date'] >= start_date) & (df['Coil_Prod_Date'] <= end_date)]

    def make_class_string(row):
        return "".join("H" if row[c] > 15 else ("L" if row[c] < 3 else "M") for c in class_cols)

    df['class_string'] = df.apply(make_class_string, axis=1)
    return df


def load_coil_signal_csv(file_path):
    return pd.read_csv(file_path)

def compute_signal_features(df, signal_features):
    results = {}
    k = 3  # threshold multiplier
    for col in signal_features:
        new_col_name = f"{col}_filtered_mean"
        if col in df.columns:
            series = df[col].dropna()
            mean = series.mean()
            std = series.std()
            lower = mean - k * std
            upper = mean + k * std
            filtered_series = series[(series >= lower) & (series <= upper)]
            results[new_col_name] = filtered_series.mean()
        else:
            results[new_col_name] = None
    return results
```

```python
class_division_df = load_class_division_csv(class_division_csv, class_cols)
coilid_to_class = dict(zip(class_division_df['Pickle ID'], class_division_df['class_string']))
```

```python
results = []
cnt = 0
for fname in os.listdir(source_folder):
    if fname.lower().endswith('.csv'):
        coil_id = fname[:-4].strip()
        signal_df = load_coil_signal_csv(os.path.join(source_folder, fname))
        row = {'Coil ID': coil_id}
        row.update(compute_signal_features(signal_df, signal_features))
        row['class_string'] = coilid_to_class.get(str(coil_id), "")
        results.append(row)
        cnt+=1
        print(f"{cnt}. Processing {coil_id}")
        
master_df = pd.DataFrame(results)
```
