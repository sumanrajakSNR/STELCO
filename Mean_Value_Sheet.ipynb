```python
import pandas as pd
import os
```

```python
excel_file = r"D:\Stelco\Work\Dynamic Correlation\QVW Data for S&R 2024-7-2025 - 1008GCQ.xlsx"
df_all_data = pd.read_excel(excel_file, sheet_name="All Data")
df_all_data.columns = df_all_data.columns.str.strip()  # Remove leading/trailing whitespace
```

```python
df_unique = df_all_data[['Pickle ID','Coil_Prod_Date']].drop_duplicates()
df_unique = df_unique.dropna(subset=['Pickle ID'])  
df_unique['Pickle ID'] = df_unique['Pickle ID'].astype(int).astype(str) 
```

```python
folder_path = r"D:\Stelco\Work\Dynamic Correlation\Key\Master Whole Phase" 
columns_to_avg = ["Operator_Q_VW_OLD_OPTECH", "Q1_Q_VW_OLD_OPTECH", "Center_Q_VW_OLD_OPTECH", 
                  "Q2_Q_VW_OLD_OPTECH", "Driver_Q_VW_OLD_OPTECH"]
```

```python
print(df_unique.columns.tolist())
```

```python
results = []
missing_files = []

for _, row in df_unique.iterrows():
    pickle_id = row['Pickle ID']
    prod_date = row['Coil_Prod_Date']
    file_path = os.path.join(folder_path, f"{pickle_id}.csv")
    
    if os.path.exists(file_path):
        try:
            print(f"Processing : {file_path}")
            df_pickle = pd.read_csv(file_path, usecols=columns_to_avg)
            avg_values = df_pickle[columns_to_avg].mean()
            results.append({
                'Pickle ID': pickle_id,
                'Coil_Prod_Date': prod_date,
                'Operator_Q_VW_Avg': avg_values["Operator_Q_VW_OLD_OPTECH"],
                'Q1_Q_VW_Avg': avg_values["Q1_Q_VW_OLD_OPTECH"],
                'Center_Q_VW_Avg': avg_values["Center_Q_VW_OLD_OPTECH"],
                'Q2_Q_VW_Avg': avg_values["Q2_Q_VW_OLD_OPTECH"],
                'Driver_Q_VW_Avg': avg_values["Driver_Q_VW_OLD_OPTECH"],
            })
        except Exception as e:
            print(f"Error processing {pickle_id}.csv: {e}")
    else:
        print(f"File not found for Pickle ID: {pickle_id}")
        missing_files.append(pickle_id)


# --- Step 3: Save the results ---
df_result = pd.DataFrame(results)
df_result.to_csv("Q_VW_Averages.csv", index=False)
print("Saved to Q_VW_Averages.csv")
```
