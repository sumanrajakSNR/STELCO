```python
import os
import glob
import pandas as pd

# Specify folder and signal list
folder_path = r"D:\Stelco\Work\Dynamic Correlation\Key\Whole_Phase_Bad"
exclude = ["Time", "Master Ramp", "Coil Length [30ms]", "Coil ID"]
output_csv = 'coil_signal_averages_bad.csv'

# Prepare output list
all_averages = []

# Iterate CSV files
for file_path in glob.glob(os.path.join(folder_path, '*.csv')):
    df = pd.read_csv(file_path)
    
    # Exclude non-signal columns
    signal_list = df.columns.difference(exclude)
    
    # Create dictionary of averages
    averages = {signal: df[signal].mean() for signal in signal_list}
    
    # Add coil ID separately
    averages['Coil ID'] = df['Coil ID'].iloc[0]  # adjust column name if needed
    
    # Append to list
    all_averages.append(averages)
```

```python
# Create DataFrame and save
out_df = pd.DataFrame(all_averages)
out_df.to_csv(output_csv, index=False)

print(f"Saved averages to {output_csv}")
```
