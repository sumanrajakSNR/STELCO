```python
# Cell 3: Define helper function
def get_nearest_indices(lengths, desired_points):
    """Find nearest indices for desired points, taking the LAST occurrence when there are duplicates"""
    indices = []
    for desired in desired_points:
        # Find all indices with minimum distance
        distances = np.abs(lengths - desired)
        min_distance = np.min(distances)
        # Get all indices that have the minimum distance
        min_indices = np.where(distances == min_distance)[0]
        # Take the LAST index (highest index number)
        indices.append(min_indices[-1])
    return indices
```

```python
# Cell 5: Initialize output settings
sample_interval = 3
output_csv = f'good_and_bad_coils_{sample_interval}_feet_sampling_combined.csv'
write_header = True

# Clear the output file if it exists
if os.path.exists(output_csv):
    os.remove(output_csv)
```

```python
# Cell 6: Main processing loop with debugging
total_coils_processed = 0
total_rows_written = 0
files_with_issues = []

for file_idx, csv_file in enumerate(csv_files):
    print(f"\n=== Processing file {file_idx + 1}/{len(csv_files)}: {os.path.basename(csv_file)} ===")
    
    try:
        df = pd.read_csv(csv_file, low_memory=False)
        print(f"Loaded {len(df)} rows")
        
        # Check if required columns exist
        missing_columns = [col for col in selected_columns if col not in df.columns]
        if missing_columns:
            print(f"WARNING: Missing columns: {missing_columns[:5]}...")  # Show first 5 missing
            # Use only available columns
            available_columns = [col for col in selected_columns if col in df.columns]
            if len(available_columns) < 3:  # Need at least Coil ID, Length, and one data column
                print(f"SKIPPING file - too few available columns ({len(available_columns)})")
                files_with_issues.append(csv_file)
                continue
        else:
            available_columns = selected_columns
            
        # Check if essential columns exist
        if 'Coil ID' not in df.columns or 'Coil Length [30ms]' not in df.columns:
            print(f"SKIPPING file - missing essential columns")
            files_with_issues.append(csv_file)
            continue
        
        unique_coils = df['Coil ID'].nunique()
        print(f"Found {unique_coils} unique coils")
        
        batch_rows = []
        coils_processed_in_file = 0
        
        for coil_id, group in df.groupby('Coil ID'):
            try:
                # Remove rows with NaN coil lengths first
                group_clean = group.dropna(subset=['Coil Length [30ms]'])
                if len(group_clean) == 0:
                    print(f"  Coil {coil_id}: SKIPPING - all coil lengths are NaN")
                    continue
                
                if len(group_clean) != len(group):
                    print(f"  Coil {coil_id}: Removed {len(group) - len(group_clean)} rows with NaN lengths")
                
                group_sorted = group_clean.sort_values('Coil Length [30ms]').reset_index(drop=True)
                start = group_sorted['Coil Length [30ms]'].min()
                stop = group_sorted['Coil Length [30ms]'].max()
                length_range = stop - start
                
                print(f"  Coil {coil_id}: {len(group)} rows, range: {length_range:.1f}m")
                
                # Skip coils with very short ranges
                if length_range < 10:
                    print(f"    SKIPPING - range too short")
                    continue
                
                desired_lengths = np.arange(start, stop + sample_interval, sample_interval)
                
                if len(desired_lengths) == 0:
                    print(f"    SKIPPING - no sampling points")
                    continue
                
                print(f"    Desired sampling points: {len(desired_lengths)}")
                
                indices = get_nearest_indices(group_sorted['Coil Length [30ms]'].values, desired_lengths)
                
                if len(indices) == 0:
                    print(f"    SKIPPING - no valid sampling points found")
                    continue
                    
                unique_indices = np.unique(indices)
                
                print(f"    Original indices: {len(indices)}, Unique indices: {len(unique_indices)}")
                
                # Debug: show some actual coil lengths
                if len(unique_indices) < 10:  # Only debug problematic coils
                    sample_lengths = group_sorted['Coil Length [30ms]'].iloc[unique_indices].values
                    print(f"    Selected lengths: {sample_lengths[:10]}")  # Show first 10
                
                # Select only available columns
                selected_data = group_sorted.iloc[unique_indices][available_columns]
                
                if len(selected_data) > 0:
                    batch_rows.append(selected_data)
                    coils_processed_in_file += 1
                    total_coils_processed += 1
                    print(f"    Added {len(selected_data)} rows")
                else:
                    print(f"    SKIPPING - no data after selection")
                    
            except Exception as e:
                print(f"    ERROR processing coil {coil_id}: {e}")
                continue
        
        # Concatenate all coils for this file
        if batch_rows:
            batch_df = pd.concat(batch_rows, ignore_index=True)
            rows_in_batch = len(batch_df)
            total_rows_written += rows_in_batch
            
            print(f"Writing {rows_in_batch} rows from {coils_processed_in_file} coils")
            batch_df.to_csv(output_csv, mode='a', header=write_header, index=False)
            write_header = False
        else:
            print(f"WARNING: No data to write from this file")
            files_with_issues.append(csv_file)
            
    except Exception as e:
        print(f"ERROR processing file: {e}")
        files_with_issues.append(csv_file)
        continue

print(f"\nAll files processed. Output saved to: {output_csv}")
```
