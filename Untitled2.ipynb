```python
import os
import pandas as pd


master_csv_path = r"D:\Stelco\Work\Dynamic Correlation\QVW Data for S&R 2024-7-2025 - 1008GCQ.xlsx"
coil_folder_path = r"D:\Stelco\Work\Dynamic Correlation\Key\Master whole Phase"
output_csv_path = "all 2988 coils to extract.csv"


df_master = pd.read_excel(master_csv_path,sheet_name='All Data')


available_ids = {
    os.path.splitext(filename)[0].strip()
    for filename in os.listdir(coil_folder_path)
    if filename.lower().endswith(".csv")
}


df_master["Pickle ID"] = pd.to_numeric(df_master["Pickle ID"], errors='coerce')
df_master = df_master.dropna(subset=["Pickle ID"])
df_master["Pickle ID"] = df_master["Pickle ID"].astype(int).astype(str)

df_master = df_master.drop_duplicates(subset="Pickle ID", keep="first")

df_filtered = df_master[df_master['Pickle ID'].isin(available_ids)]

df_filtered.to_csv(output_csv_path, index=False)

print(f"✅ Filtered CSV saved with {len(df_filtered)} rows at: {output_csv_path}")
```

```python
import pandas as pd
import math
import os

# Path to the filtered CSV
input_csv = "all 2988 coils to extract.csv"

# Output directory to save parts
output_dir = "Seperated All Coils/"
os.makedirs(output_dir, exist_ok=True)

# Read the full CSV
df = pd.read_csv(input_csv)

# Total rows and number of parts
total_rows = len(df)
rows_per_file = 500
num_parts = math.ceil(total_rows / rows_per_file)

# Split and save each part
for i in range(num_parts):
    start = i * rows_per_file
    end = start + rows_per_file
    df_part = df.iloc[start:end]
    
    part_filename = os.path.join(output_dir, f"filtered_part_{i+1}.csv")
    df_part.to_csv(part_filename, index=False)
    print(f"✅ Saved: {part_filename} ({len(df_part)} rows)")
```
