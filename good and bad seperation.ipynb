```python
import pandas as pd

# Lists of coil IDs
bad_coils = [5514681,5514677,5514672,5514670,5514668,5514674,5513807,5512575,5503973,5503968,5504079,5499471,5522770,5520684,5519680,5520680,5519376]
good_coils = [5484972,5484981,5485581,5487008,5488171,5486250,5487948,5487932,5487937,5487949,5487953,5487951,5490707,5492525,5493226,5493229,5493231,5493809,5499738,5499642,5499278,5500841,5504100,5504098,5504094,5510195]

sample_interval = 3
```

```python
def split_coil_csv(input_filepath):
    df = pd.read_csv(input_filepath)

    # Filter for bad and good coils
    df_bad = df[df['Coil ID'].isin(bad_coils)]
    df_good = df[df['Coil ID'].isin(good_coils)]

    # Save to separate CSV files
    df_bad.to_csv(f'bad_coils_{sample_interval}_feet_sampling_combined_added_gauge_target_phase_division.csv', index=False)
    df_good.to_csv(f'good_coils_{sample_interval}_feet_sampling_combined_added_gauge_target_phase_division.csv', index=False)

    return df_bad, df_good

# Example usage:
bad_df, good_df = split_coil_csv(f"good_and_bad_coils_{sample_interval}_feet_sampling_combined_added_gauge_target_phase_division.csv")
# print(f"Bad coils rows: {len(bad_df)}")
# print(f"Good coils rows: {len(good_df)}")
```

## Adding Coil Status Column

```python
def label_coil(coil_id):
    if coil_id in good_coils:
        return 'Good'
    elif coil_id in bad_coils:
        return 'Bad'
    else:
        return 'Unknown'

# Specify your Excel file and sheet name
filename = "Good_and_Bad_Coils_Phasewise_correlation_summary_with_comparison_15-90.xlsx"
sheet = 'Good_and_Bad_Coils_Phasewise_co'

# Read the file
df = pd.read_excel(filename, sheet_name=sheet)

# Add the column
df['Coil Status'] = df['Coil ID'].apply(label_coil)

# Update the file in-place
with pd.ExcelWriter(filename, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    df.to_excel(writer, sheet_name=sheet, index=False)
```
