```python
# Step 1: Aggregate signal averages from each CSV
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        df = pd.read_csv(os.path.join(folder_path, filename))
        means = {}
        for sig in signal_cols:
            if sig in df.columns:
                means[sig] = df[sig].mean()
            else:
                means[sig] = np.nan
        means['coil_id'] = filename.replace('.csv', '')
        coil_data.append(means)

combined_df = pd.DataFrame(coil_data).set_index('coil_id')

# Step 2: Compute correlation matrix
corr_matrix = combined_df.corr().abs()

# Step 3: Identify and drop highly correlated signals
threshold = 0.9
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(bool))
to_drop = [col for col in upper.columns if any(upper[col] > threshold)]
filtered_df = combined_df.drop(columns=to_drop)

print("Signals retained after correlation filtering:", filtered_df.columns.tolist())
```

```python
cols.append(filtered_df.columns.tolist())
```

```python
with open('signals.txt', 'w') as file:
    for item in cols:
        file.write(f"\n{item}\n")
```
