```python
import os

folder1_path = r"D:\Stelco\Work\Dynamic Correlation\Key\Whole_Phase_Good_Clustered"
folder2_path = r"D:\Stelco\Work\Dynamic Correlation\Key\Whole_Phase_Good_Clustered_Filtered"

def list_unique_files(folder1, folder2):
    files_folder1 = set(os.listdir(folder1))
    files_folder2 = set(os.listdir(folder2))

    unique_files = files_folder1 - files_folder2
    return list(unique_files)

unique_files_list = list_unique_files(folder1_path, folder2_path)

print("Files in folder1 but not in folder2:")
for filename in unique_files_list:
    print(filename)
```

```python
import shutil

source_folder = r"D:\Stelco\Work\Dynamic Correlation\Key\Whole_Phase_Good_Clustered"
destination_folder = r"D:\Stelco\Work\Dynamic Correlation\Key\Whole_Phase_Bad_Clustered"

# Make sure destination folder exists, create if not
os.makedirs(destination_folder, exist_ok=True)

def copy_files(file_list, src_folder, dest_folder):
    for filename in file_list:
        src_path = os.path.join(src_folder, filename)
        dest_path = os.path.join(dest_folder, filename)
        try:
            shutil.copy2(src_path, dest_path)
            print(f"Copied {filename} to {dest_folder}")
        except Exception as e:
            print(f"Failed to copy {filename}: {e}")

copy_files(unique_files_list, source_folder, destination_folder)
```
