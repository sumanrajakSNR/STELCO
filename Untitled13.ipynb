```python
import pandas as pd

# Load CSV file
df = pd.read_csv('bucket_coil_signal_averages_allsignals_phasewise_rollingmode_reshaped_with_clusters.csv')

# Filter rows with Bucket_ID = 12
df_filtered = df[df['Bucket_ID'] == 12]

signals = ['Stand 4 Top Current Feedback', 'Stand 1 Bottom Current Feedback', 'Stand 1 Roll Force Increase Limit (based on predicted run force)', 'Stand 3 Top Current Feedback', 'Stand 2 Top Current Feedback', 'Stand 2 Bottom Current Feedback', 'Stand 3 Drive Speed Feedback', 'Stand 3 Bottom Current Feedback', 'Stand 4 Bottom Current Feedback', 'Stand 4 - Operator Side Force']
phases = [1, 2, 3]
signal_columns = [f"{sig}_Phase{p}" for sig in signals for p in phases]

# Select only Bucket_ID and the desired signal columns
df_selected = df_filtered[['Bucket_ID','rollingmode','cluster_label'] + signal_columns]
```

```python
df_selected = df_selected[(df_selected['rollingmode'] == 'Sheet')]
```

```python
df_selected.to_csv("sheetdemo.csv",index=False)
```
