```python
import pandas as pd
```

```python
# Load CSV files
df1 = pd.read_csv("anomalies_outside_threshold_volatility.csv")   # Contains coil ids
df2 = pd.read_csv("coil_qvw_averages.csv")  # Contains coil ids and 5 qvw averages columns
```

```python
# Extract list of coil ids from first df
coil_ids = df1['Coil ID'].unique()

# Filter second df to matching coil ids
df_filtered = df2[df2['Coil ID'].isin(coil_ids)]

# List of qvw columns (adjust column names as needed)
qvw_columns = ["Center_Q_VW_OLD_OPTECH","Driver_Q_VW_OLD_OPTECH","Operator_Q_VW_OLD_OPTECH","Q1_Q_VW_OLD_OPTECH","Q2_Q_VW_OLD_OPTECH"]

records = []

# Iterate filtered rows to check qvw averages > 6
for _, row in df_filtered.iterrows():
    for col in qvw_columns:
        if row[col] > 6:
            records.append({
                'Coil ID': row['Coil ID'],
                'qvw_column': col,
                'qvw_value': row[col]
            })
```

```python
# Create dataframe and export
result_df = pd.DataFrame(records)
result_df.to_csv('coils_qvw_over_6_volatility.csv', index=False)
```
