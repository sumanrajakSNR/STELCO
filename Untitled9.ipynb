```python
import pandas as pd

# Load the CSV with clusters
df_clusters = pd.read_csv('coils_clustering_data_badcoils.csv')  # CSV containing "Coil ID" and "K means Cluster"

# Load the other CSV you want to add cluster info to
df_other = pd.read_csv('coil_anomaly_isolationforest_badcoils.csv')  # CSV containing "Coil ID" and other columns

# Merge on "Coil ID" to get cluster column in other dataframe
df_merged = pd.merge(df_other, df_clusters[['Coil ID', 'K means Cluster']], on='Coil ID', how='inner')

df_merged.dropna()

# Save the resulting CSV with cluster info added
df_merged.to_csv('coil_anomaly_isolationforest_badcoils_clustering_data.csv', index=False)
```
