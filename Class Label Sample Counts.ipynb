```python
import pandas as pd
file_path = "new_class_division_coils_without_M.csv"
output_path = "new_Class_Label_counts_without_M.csv"
# Read the CSV file
# Replace 'input.csv' with your file name
df = pd.read_csv(file_path)

# Count occurrences of each class label
# Replace 'Class Label' with the actual column name in your file
class_counts = df['Class Label'].value_counts().reset_index()

# Rename columns for clarity
class_counts.columns = ['Class Label', 'No of Samples']

# Save to a new CSV
class_counts.to_csv(output_path, index=False)

print(f"Class label counts saved to {output_path}")
```
