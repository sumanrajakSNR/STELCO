```python
import pandas as pd

# Define the file path
file_path = r"D:\Stelco\Work\Dynamic Correlation\QVW Data for S&R 2024-7-2025 - 1008GCQ.xlsx"

# Define the relevant columns
class_cols = ['OPE Avg QVW', 'Q1 Avg QVW', 'Center Avg QVW', 'Q2 Avg QVW', 'DRE Avg QVW']

# Read the CSV file
df = pd.read_excel(file_path,sheet_name="All Data")

# Function to assign H, M, or L per row
def assign_class_label(row):
    return "".join("H" if row[c] > 15 else ("L" if row[c] < 3 else "M") for c in class_cols)

# Apply the logic and create the new column
df['Class Label'] = df.apply(assign_class_label, axis=1)

# Create the new DataFrame
output_df = df[['Pickle ID', 'Class Label'] + class_cols]

# Save to new CSV
output_df.to_csv('new_class_division_coils.csv', index=False)
```
