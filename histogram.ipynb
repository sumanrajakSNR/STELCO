```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\USER\Desktop\Jupyter Notebook\Stelco\Histogram Data.csv")

plt.hist(df['Contribution Factor %'], bins=20, color='skyblue', edgecolor='black')

# Force integer ticks on the y-axis
max_count = plt.gca().get_ylim()[1]
plt.yticks(np.arange(0, int(max_count) + 1, 1))

plt.title('Contribution Factor % | Exit Tension with Driver | Phase 2 | >15%')
plt.xlabel('Contribution Factor %')
plt.ylabel('Frequency of Coils')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
