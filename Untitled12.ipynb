```python
import pandas as pd
import os

# Assign phase function you provided
def assign_phase_column_by_coil(df, coil_col, ramp_col):
    def assign_phase(group):
        if group[ramp_col].max() <= 15:
            group['Phase'] = None
            return group
        norm_val = group[ramp_col].mode()[0]
        norm_df = group[group[ramp_col] == norm_val]
        if norm_df.empty:
            group['Phase'] = None
            return group
        phase1_end = norm_df.index[0]
        phase2_end = norm_df.index[-1]
        group['Phase'] = None
        group.loc[:phase1_end, 'Phase'] = 'Phase 1'
        group.loc[phase1_end:phase2_end, 'Phase'] = 'Phase 2'
        group.loc[phase2_end+1:, 'Phase'] = 'Phase 3'
        return group

    return df.groupby(coil_col).apply(assign_phase).reset_index(drop=True)

# Paths
merged_folder = r"D:\Stelco\Work\Dynamic Correlation\Key\Whole_Phase_Bad_Merged"
phase_output_folder = r"D:\Stelco\Work\Dynamic Correlation\Key\Whole_Phase_Bad_Phased"

# Ensure output folder exists
os.makedirs(phase_output_folder, exist_ok=True)

# Process each merged CSV for phase assignment
for filename in os.listdir(merged_folder):
    if filename.endswith('.csv'):
        file_path = os.path.join(merged_folder, filename)
        df = pd.read_csv(file_path)
        
        # Apply phase assignment, use column names as per your data: 'Coil ID' and 'Master Ramp'
        df_with_phase = assign_phase_column_by_coil(df, coil_col='Coil ID', ramp_col='Master Ramp')
        
        # Save updated file to new folder
        save_path = os.path.join(phase_output_folder, filename)
        df_with_phase.to_csv(save_path, index=False)
```

```python
import pandas as pd
import os

coil_folder = r"D:\Stelco\Work\Dynamic Correlation\Key\Whole_Phase_Good_Clustered"
clustering_csv = r"D:\Stelco\Work\Dynamic Correlation\cluster.csv"
output_csv = r"D:\Stelco\Work\Dynamic Correlation\cluster_merged.csv"

# Only these columns to extract
columns_to_extract = ['CP_X4GAUGE', 'Reduction']

# Load clustering CSV
clustering_df = pd.read_csv(clustering_csv)

summary_data = []

for filename in os.listdir(coil_folder):
    if filename.endswith('.csv'):
        coil_id = os.path.splitext(filename)[0]
        file_path = os.path.join(coil_folder, filename)
        df = pd.read_csv(file_path)
        
        coil_summary = {'Coil ID': int(coil_id)}
        for col in columns_to_extract:
            if col in df.columns:
                coil_summary[col] = df[col].mode()[0] if not df[col].mode().empty else df[col].iloc[0]
            else:
                coil_summary[col] = None
        summary_data.append(coil_summary)

extracted_df = pd.DataFrame(summary_data)

# Merge extracted columns with clustering CSV on Coil ID
merged_df = clustering_df.merge(extracted_df, on='Coil ID', how='left')

# Save the enriched CSV
merged_df.to_csv(output_csv, index=False)
```
