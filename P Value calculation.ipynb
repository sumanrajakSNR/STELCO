```python
import pandas as pd
from scipy.stats import t
import os

def correlation_p_value(r, n):
    t_stat = r * (n - 2) ** 0.5 / (1 - r ** 2) ** 0.5
    p_val = 2 * t.sf(abs(t_stat), n - 2)
    return p_val

# Replace with your actual data loading
df = pd.read_csv('Good_and_Bad_Coils_Phasewise_correlation_summary.csv')
```

## not unique

```python
# Convert contribution factor % to decimal correlation
df['Correlation'] = df['Contribution Factor %'] / 100

# Calculate p-values
df['p_value'] = df.apply(lambda row: correlation_p_value(row['Correlation'], row['Total Windows']), axis=1)

# Select columns to save
output_df = df[['Signal Feature', 'Shape Feature', 'Phase','Contribution Factor %', 'p_value']]

output_csv = 'correlation_pvalues.csv'

if os.path.exists(output_csv):
    os.remove(output_csv)
    
# Save to CSV file
output_df.to_csv(output_csv, index=False)

print(output_df.head())
```
